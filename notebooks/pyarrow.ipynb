```python
from mountainash_data.dataframes import DataFrameUtils
from mountainash_constants import CONST_DATAFRAME_FRAMEWORK
import pyarrow as pa


dataDictExample1 = {
    "column1": [1, 2, 3],
    "column2": [4, 5, 6],
    "column3": ["A", "B", "C"]
}

df_pyarrow_table: pa.Table = DataFrameUtils.create_dataframe(CONST_DATAFRAME_FRAMEWORK.PYARROW_TABLE.value, dataDictExample1)

df_pyarrow_table
```

```python
pab = df_pyarrow_table.to_batches(max_chunksize=2)
type(pab[0])
```

```python
record: pa.RecordBatch = pab[0]
record

record.get_total_buffer_size()
```

```python
record_field = "column1"
values = {}

if record_field in record.column_names:
    # If the field exists in the record, get its value
    field_array = record[record_field]
    if len(field_array) > 0:
        values[record_field] = field_array[1].as_py()

values
```
